1. For every year, model the data as a multinomial probability distribution for the fol-lowing features:  (1) gender, (2) age, (3) purpose of visit. 
    -  Do it overall as well as country-wiseand continent-wise.
3. Find the years when the entropy of the distributions (three seaparate) for the overall count is themaximum and the minimum.
    - Repeat it for every country and continent.

```python
# import pandas as pd

# Load data
# df = pd.read_csv("your_dataset.csv")

# Define features with corresponding columns for each year
years = [2013, 2014, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
gender_columns = {year: [f'MALE {year}', f'FEMALE {year}'] for year in years}
age_columns = {year: [f'0-14 {year}', f'15-24 {year}', f'25-34 {year}', f'35-44 {year}', f'45-54 {year}', f'55-64 {year}', f'65 AND ABOVE {year}'] for year in years}
purpose_columns = {year: [f'BUSINESS AND PROFESSIONAL {year}', f'LEISURE HOLIDAY AND RECREATION {year}', f'MEDICAL {year}', f'INDIAN DIASPORA {year}', f'OTHERS {year}'] for year in years}

# Prepare lists to store probabilities for each grouping
results = []

def calculate_probabilities(df, columns):
    """Calculate multinomial probabilities for given columns in a DataFrame, excluding zero-total rows."""
    # Filter out rows with a total of zero
    df = df[df[columns].sum(axis=1) > 0]
    total_counts = df[columns].sum(axis=1)
    probabilities = df[columns].div(total_counts, axis=0)
    return probabilities

# Loop through each year to process gender, age, and purpose of visit data
for year in years:
    # Overall probabilities for each feature
    gender_probs = calculate_probabilities(df, gender_columns[year])
    age_probs = calculate_probabilities(df, age_columns[year])
    purpose_probs = calculate_probabilities(df, purpose_columns[year])
    
    # Add probabilities to results for overall distribution
    results.append({
        'YEAR': year,
        'LEVEL': 'Overall',
        'GENDER_PROBS': gender_probs.mean().to_dict(),
        'AGE_PROBS': age_probs.mean().to_dict(),
        'PURPOSE_PROBS': purpose_probs.mean().to_dict()
    })
    
    # Country-wise probabilities
    for country, group in df.groupby('COUNTRY OF NATIONALITY'):
        gender_probs = calculate_probabilities(group, gender_columns[year])
        age_probs = calculate_probabilities(group, age_columns[year])
        purpose_probs = calculate_probabilities(group, purpose_columns[year])
        
        results.append({
            'YEAR': year,
            'LEVEL': f'Country - {country}',
            'GENDER_PROBS': gender_probs.mean().to_dict(),
            'AGE_PROBS': age_probs.mean().to_dict(),
            'PURPOSE_PROBS': purpose_probs.mean().to_dict()
        })
        
    # Continent-wise probabilities
    for continent, group in df.groupby('CONTINENT'):
        gender_probs = calculate_probabilities(group, gender_columns[year])
        age_probs = calculate_probabilities(group, age_columns[year])
        purpose_probs = calculate_probabilities(group, purpose_columns[year])
        
        results.append({
            'YEAR': year,
            'LEVEL': f'Continent - {continent}',
            'GENDER_PROBS': gender_probs.mean().to_dict(),
            'AGE_PROBS': age_probs.mean().to_dict(),
            'PURPOSE_PROBS': purpose_probs.mean().to_dict()
        })

# Convert results to DataFrame for better readability
results_df = pd.DataFrame(results)

# Separate DataFrames for Country and Continent levels
country_df = results_df[results_df['LEVEL'].str.startswith('Country')].reset_index(drop=True)
continent_df = results_df[results_df['LEVEL'].str.startswith('Continent')].reset_index(drop=True)
overall_df = results_df[results_df['LEVEL'].str.startswith('Overall')].reset_index(drop=True)

# Display the two DataFrames
# print("Country-wise Probabilities DataFrame:")
# print(country_df)
# print("\nContinent-wise Probabilities DataFrame:")
# print(continent_df)
# country_df.head(10)
```

```python
import numpy as np
from scipy.stats import entropy

# Load the data
# df = pd.read_csv('your_data.csv')
# df=df

# Columns for each distribution (Gender, Age, Purpose)
gender_columns = [col for col in df.columns if 'MALE' in col or 'FEMALE' in col]
age_columns = [col for col in df.columns if any(age in col for age in ['0-14', '15-24', '25-34', '35-44', '45-54', '55-64', '65 AND ABOVE'])]
purpose_columns = [col for col in df.columns if any(purpose in col for purpose in ['BUSINESS AND PROFESSIONAL', 'LEISURE HOLIDAY AND RECREATION', 'MEDICAL', 'INDIAN DIASPORA', 'OTHERS'])]

# Prepare the results list
results = []

# Extract unique years from the column names
years = sorted(set([col.split()[-1] for col in gender_columns]))  # Extract year from 'MALE 2013', 'FEMALE 2013'

for year in years:
    # Define year-specific columns
    gender_cols = [f'MALE {year}', f'FEMALE {year}']
    age_cols = [f'0-14 {year}', f'15-24 {year}', f'25-34 {year}', f'35-44 {year}', f'45-54 {year}', f'55-64 {year}', f'65 AND ABOVE {year}']
    purpose_cols = [f'BUSINESS AND PROFESSIONAL {year}', f'LEISURE HOLIDAY AND RECREATION {year}', f'MEDICAL {year}', f'INDIAN DIASPORA {year}', f'OTHERS {year}']

    # Drop rows with NaN values in any of the relevant columns
    year_gender_data = df[gender_cols].dropna(axis=0, how='any')
    year_age_data = df[age_cols].dropna(axis=0, how='any')
    year_purpose_data = df[purpose_cols].dropna(axis=0, how='any')

    if not year_gender_data.empty and not year_age_data.empty and not year_purpose_data.empty:
        # Overall entropy
        overall_gender_entropy = entropy(year_gender_data.sum())
        overall_age_entropy = entropy(year_age_data.sum())
        overall_purpose_entropy = entropy(year_purpose_data.sum())

        # Append overall entropies
        results.append({
            'Year': year,
            'Level': 'Overall',
            'Gender_Entropy': overall_gender_entropy,
            'Age_Entropy': overall_age_entropy,
            'Purpose_Entropy': overall_purpose_entropy
        })

        # Country-wise entropy
        for country in df['COUNTRY OF NATIONALITY'].unique():
            country_data = df[df['COUNTRY OF NATIONALITY'] == country].dropna(axis=0, how='any')

            if not country_data.empty:
                country_gender_entropy = entropy(country_data[gender_cols].sum())
                country_age_entropy = entropy(country_data[age_cols].sum())
                country_purpose_entropy = entropy(country_data[purpose_cols].sum())

                results.append({
                    'Year': year,
                    'Level': f'Country - {country}',
                    'Gender_Entropy': country_gender_entropy,
                    'Age_Entropy': country_age_entropy,
                    'Purpose_Entropy': country_purpose_entropy
                })

        # Continent-wise entropy
        for continent in df['CONTINENT'].unique():
            continent_data = df[df['CONTINENT'] == continent].dropna(axis=0, how='any')

            if not continent_data.empty:
                continent_gender_entropy = entropy(continent_data[gender_cols].sum())
                continent_age_entropy = entropy(continent_data[age_cols].sum())
                continent_purpose_entropy = entropy(continent_data[purpose_cols].sum())

                results.append({
                    'Year': year,
                    'Level': f'Continent - {continent}',
                    'Gender_Entropy': continent_gender_entropy,
                    'Age_Entropy': continent_age_entropy,
                    'Purpose_Entropy': continent_purpose_entropy
                })

# Convert results to DataFrame
entropy_df = pd.DataFrame(results)

# Debug: Print the columns and first few rows of entropy_df
print("entropy_df columns:", entropy_df.columns)
# print(entropy_df.head())
# entropy_df.head()
```

- Continent-wise Entropy

```python
# import pandas as pd
# import numpy as np
# from scipy.stats import entropy

# Load data
# df = pd.read_csv("your_dataset.csv")

# Define features with corresponding columns for each year
years = [2013, 2014, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
gender_columns = {year: [f'MALE {year}', f'FEMALE {year}'] for year in years}
age_columns = {year: [f'0-14 {year}', f'15-24 {year}', f'25-34 {year}', f'35-44 {year}', f'45-54 {year}', f'55-64 {year}', f'65 AND ABOVE {year}'] for year in years}
purpose_columns = {year: [f'BUSINESS AND PROFESSIONAL {year}', f'LEISURE HOLIDAY AND RECREATION {year}', f'MEDICAL {year}', f'INDIAN DIASPORA {year}', f'OTHERS {year}'] for year in years}

# Prepare lists to store entropies for each grouping
continent_results = []

# Define a function to calculate entropy safely
def calculate_entropy(df, columns):
    """Calculate entropy of the sum of the given columns in a DataFrame, handling NaN values."""
    # Filter out rows with a total of zero
    df = df[df[columns].sum(axis=1) > 0]
    total_counts = df[columns].sum()
    if total_counts.sum() > 0:  # Ensure there is data to calculate entropy
        probabilities = total_counts / total_counts.sum()
        return entropy(probabilities)
    return np.nan

# Loop through each year to process gender, age, and purpose of visit data for continents
for year in years:
    # Continent-wise entropy calculations
    for continent, group in df.groupby('CONTINENT'):
        # Calculate entropies for gender, age, and purpose
        gender_entropy = calculate_entropy(group, gender_columns[year])
        age_entropy = calculate_entropy(group, age_columns[year])
        purpose_entropy = calculate_entropy(group, purpose_columns[year])
        
        # Store results in the list
        continent_results.append({
            'Year': year,
            'Continent': continent,
            'Gender_Entropy': gender_entropy,
            'Age_Entropy': age_entropy,
            'Purpose_Entropy': purpose_entropy
        })

# Convert continent results to DataFrame
continentwise_entropy_df = pd.DataFrame(continent_results)

# Display the continent entropy results
print("Continent-wise Yearly Entropy Analysis:")
# print(continentwise_entropy_df)
# continentwise_entropy_df.head(10)
```

```python
# Finding overall min and max for Gender, Age, and Purpose entropy across all continents and years
overall_min_max_results = []

for feature in ['Gender_Entropy', 'Age_Entropy', 'Purpose_Entropy']:
    # Overall maximum
    max_row = continentwise_entropy_df.loc[continentwise_entropy_df[feature].idxmax()]
    overall_max = {
        'Feature': feature,
        'Type': 'Max',
        'Year': max_row['Year'],
        'Continent': max_row['Continent'],
        'Entropy_Value': max_row[feature]
    }
    overall_min_max_results.append(overall_max)
    
    # Overall minimum
    min_row = continentwise_entropy_df.loc[continentwise_entropy_df[feature].idxmin()]
    overall_min = {
        'Feature': feature,
        'Type': 'Min',
        'Year': min_row['Year'],
        'Continent': min_row['Continent'],
        'Entropy_Value': min_row[feature]
    }
    overall_min_max_results.append(overall_min)

# Convert to DataFrame for better readability
overall_continentwise_min_max_df = pd.DataFrame(overall_min_max_results)

# Display the overall min and max results
print("Overall Min and Max Entropy for Each Feature across Continents:")
# print(overall_continentwise_min_max_df)
# overall_continentwise_min_max_df
```
