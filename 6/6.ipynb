```python
# Assuming dvf_df is your original DataFrame
df = dvf_df.copy()

# Check the columns in your DataFrame
columns_in_df = df.columns
# print("Columns available in the DataFrame:", columns_in_df)

# Dynamically filter out valid year columns (excluding 2015)
valid_years = [str(year) for year in range(2013, 2023) if f'DOMESTIC_{year}' in columns_in_df and f'FOREIGN_{year}' in columns_in_df]

# List of year columns (DOMESTIC and FOREIGN for each available year)
year_columns = []
for year in valid_years:
    year_columns.append(f'DOMESTIC_{year}')
    year_columns.append(f'FOREIGN_{year}')

# Calculate the total sum for each year (sum of domestic and foreign for each year)
total_per_year = df[year_columns].sum()

# Check if there are any zero or NaN values in the total sums (which could lead to NaNs in the probability calculation)
# print("Total sum for each year:\n", total_per_year)

# Handle any NaN or zero values in total_per_year by replacing them with a small number or skipping them
total_per_year = total_per_year.replace(0, 1)  # Avoid division by zero by replacing zeros with 1
total_per_year = total_per_year.fillna(1)  # Replace NaN with 1 to avoid NaN in division

# Create a new DataFrame for probabilities (don't overwrite original df)
probabilities_dvf_df = df[['STATE/UT']].copy()  # Ensure STATE/UT remains as a column, not index

# Calculate the probabilities for each row (state/UT) by dividing the counts by the total sum for each year
for year in valid_years:  # Looping through the valid years
    for domestic_foreign in ['DOMESTIC', 'FOREIGN']:
        column = f'{domestic_foreign}_{year}'
        # Ensure there are no NaN or infinite values before dividing
        probabilities_dvf_df[column] = df[column] / total_per_year[column]

# Calculate the overall probability for each state (sum of probabilities across all years for each state)
# Sum probabilities for both DOMESTIC and FOREIGN columns for each state
probabilities_dvf_df['overall_probability'] = 0  # Initialize a new column for overall probability

# Iterate over each row (state/UT) and calculate the overall probability for that state
for index, row in probabilities_dvf_df.iterrows():
    total_prob = row[year_columns].sum()  # Sum over all DOMESTIC and FOREIGN columns for each state
    probabilities_dvf_df.at[index, 'overall_probability'] = total_prob / len(year_columns)  # Normalize by the number of columns

# Reorder columns to make 'overall_probability' the second column (STATE/UT first, overall_probability second)
cols = ['STATE/UT', 'overall_probability'] + [col for col in probabilities_dvf_df.columns if col not in ['STATE/UT', 'overall_probability']]
probabilities_dvf_df = probabilities_dvf_df[cols]

# Check the resulting probabilities_dvf_df for NaNs or issues
# print("Calculated Probabilities with Overall Probability as Second Column:\n", probabilities_dvf_df)

# Calculate overall probabilities for the entire dataset (sum of all DOMESTIC and FOREIGN columns)
overall_probabilities = df[year_columns].sum() / total_per_year.sum()

# Adding the overall probabilities as a row to the dataframe (not an index)
probabilities_dvf_df.loc['Overall'] = overall_probabilities

# Display the probabilities
# print(probabilities_dvf_df)
probabilities_dvf_df
```

```python
# import pandas as pd
import numpy as np
from scipy.stats import entropy

# Sample data - replace with the actual 'dvf_df' DataFrame
# dvf_df = pd.read_csv('path_to_your_data.csv')  # Load data if from CSV

# Column pairs for domestic and foreign tourist counts per year
years = ["2013", "2014", "2016", "2017", "2018", "2019", "2020", "2021", "2022"]
columns = [(f"DOMESTIC_{year}", f"FOREIGN_{year}") for year in years]

# List to store results
results = []

# Calculate entropy for each state and find the max/min entropy years
for i, row in dvf_df.iterrows():
    state = row["STATE/UT"]
    entropies = {}
    
    # Calculate entropy for each year
    for year, (dom_col, for_col) in zip(years, columns):
        # Get the domestic and foreign counts for the current year
        domestic_count = row[dom_col]
        foreign_count = row[for_col]
        
        # Calculate the entropy for the domestic count
        if domestic_count > 0:
            domestic_prob = domestic_count / (domestic_count + foreign_count)
            ent_dom = -domestic_prob * np.log2(domestic_prob) if domestic_prob > 0 else 0
        else:
            ent_dom = np.nan
        
        # Calculate the entropy for the foreign count
        if foreign_count > 0:
            foreign_prob = foreign_count / (domestic_count + foreign_count)
            ent_for = -foreign_prob * np.log2(foreign_prob) if foreign_prob > 0 else 0
        else:
            ent_for = np.nan
        
        # Store the entropy values for each year (separately for domestic and foreign)
        entropies[f"Domestic_Entropy_{year}"] = ent_dom
        entropies[f"Foreign_Entropy_{year}"] = ent_for
    
    # Add the row to results list
    results.append({
        "STATE/UT": state,
        **entropies
    })

# Convert results list to a DataFrame
entropy_results = pd.DataFrame(results)

# Now calculate overall max and min entropy for domestic and foreign separately
# For domestic entropy
domestic_entropy_columns = [col for col in entropy_results.columns if 'Domestic_Entropy' in col]
foreign_entropy_columns = [col for col in entropy_results.columns if 'Foreign_Entropy' in col]

# Find the overall max and min entropy for domestic (across all rows and all years)
overall_max_domestic_entropy = entropy_results[domestic_entropy_columns].max().max()
overall_min_domestic_entropy = entropy_results[domestic_entropy_columns].min().min()

# Find the overall max and min entropy for foreign (across all rows and all years)
overall_max_foreign_entropy = entropy_results[foreign_entropy_columns].max().max()
overall_min_foreign_entropy = entropy_results[foreign_entropy_columns].min().min()

# Display the result
# print("\nOverall Maximum and Minimum Entropy (Domestic and Foreign):")
# print(f"Overall Maximum Domestic Entropy: {overall_max_domestic_entropy}")
# print(f"Overall Minimum Domestic Entropy: {overall_min_domestic_entropy}")
# print(f"Overall Maximum Foreign Entropy: {overall_max_foreign_entropy}")
# print(f"Overall Minimum Foreign Entropy: {overall_min_foreign_entropy}")
overall_entropy_df = pd.DataFrame({
    "Category": ["Domestic", "Foreign"],
    "Max_Entropy": [overall_max_domestic_entropy, overall_max_foreign_entropy],
    "Min_Entropy": [overall_min_domestic_entropy, overall_min_foreign_entropy]
})
```
